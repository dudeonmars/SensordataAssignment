```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("acc_data.csv")
df.shape
```

```python
plt.figure(figsize=(10, 10))
plt.title('Sccelerometer data over time')
ax = plt.gca()
ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
plt.subplot(3, 1, 1)
plt.plot(df['ms'], df['y'], color='red')
plt.ylabel('y-axis')
plt.subplot(3, 1, 2)
plt.plot(df['ms'], df['x'], color='green')
plt.ylabel('x-asis')
plt.subplot(3, 1, 3)
plt.plot(df['ms'], df['z'])
plt.xlabel('ms')
plt.ylabel('z-axis')
plt.show()
```
